```python
import numpy as np
import pandas as pd
```

```python
#Initial scrape of data

url = "https://fbref.com/en/comps/9/"
seasons = ['schedule/Premier-League-Scores-and-Fixtures'
          , '2022-2023/schedule/2022-2023-Premier-League-Scores-and-Fixtures'
          , '2021-2022/schedule/2021-2022-Premier-League-Scores-and-Fixtures'
          , '2020-2021/schedule/2020-2021-Premier-League-Scores-and-Fixtures'
          , '2019-2020/schedule/2019-2020-Premier-League-Scores-and-Fixtures'
          , '2018-2019/schedule/2018-2019-Premier-League-Scores-and-Fixtures'
          , '2017-2018/schedule/2017-2018-Premier-League-Scores-and-Fixtures']


def get_match_data(seasons):
    result_df = []
    for i in range(len(seasons)):
        page = url + seasons[i]
        html = pd.read_html(page, header=0)
        df = html[0]
        raw = df.reset_index(drop=True)
        games = raw
        result_df.append(games)
    
    return result_df
result_df = get_match_data(seasons)
```

```python
#Dropping unnecessary columns and null values.
columns_to_keep = ['Home', 'Away', 'xG', 'xG.1', 'Score']

for i in range(len(result_df)):
    result_df[i] = result_df[i][columns_to_keep]

for i in range(len(result_df)):
    result_df[i] = result_df[i].dropna()
```

```python
# Parse score column to get winners
def get_winner(score):
    print("Processing score:", score)
    score = str(score)
    parts = score.split(score[1])
    print("Parts after split:", parts)
    if len(parts) == 2:
        score_home, score_away = map(int, parts)
        if score_home > score_away:
            return 'Home'
        elif score_home < score_away:
            return 'Away'
        else:
            return 'Draw'
    else:
        return 'Invalid Score'

for df in result_df:
    df['Winner'] = df['Score'].apply(get_winner)
    print(df)  # Print the dataframe after applying the function

result_df[0]
```

```python
#Scrape more data to clean and add to df1

urlend = ['Premier-League-Stats'
         , '2022-2023/2022-2023-Premier-League-Stats'
         , '2021-2022/2021-2022-Premier-League-Stats'
         , '2020-2021/2020-2021-Premier-League-Stats'
         , '2019-2020/2019-2020-Premier-League-Stats'
         , '2018-2019/2018-2019-Premier-League-Stats'
         , '2017-2018/2017-2018-Premier-League-Stats']

def get_standings(urlend):
    standings_dfs = []
    for i in range(len(urlend)):
        site = url + urlend[i]
        html2 = pd.read_html(site, header=0)
        df2 = html2[0]
        raw2 = df2.reset_index(drop=True)
        raw2 = raw2.fillna(0)
        teams = raw2
        standings_dfs.append(teams)
        
    return standings_dfs

standings_dfs = get_standings(urlend)
standings_dfs
```

```python
#Fill in values/columns in df1 with values from standings table
for i in range(len(result_df)):
    result_df[i]['Home_Rank'] = result_df[i]['Home'].map(standings_dfs[i].set_index('Squad')['Rk'])
    result_df[i]['Away_Rank'] = result_df[i]['Away'].map(standings_dfs[i].set_index('Squad')['Rk'])
    result_df[i]['Home_GD90'] = result_df[i]['Home'].map(standings_dfs[i].set_index('Squad')['xGD/90'])
    result_df[i]['Away_GD90'] = result_df[i]['Away'].map(standings_dfs[i].set_index('Squad')['xGD/90'])
```

```python
#Clean up the appearance a bit
for i in range(len(result_df)):
    result_df[i] = result_df[i][['Home', 'Away', 'xG', 'xG.1', 'Home_GD90', 'Away_GD90', 'Home_Rank', 'Away_Rank', 'Winner']]
result_df[0]
```

```python
#Combine all df's in result_df[] and save to .csv
final_df = pd.concat(result_df, ignore_index=True)

final_df = final_df.to_csv('stats_data2.csv', index=False)
```
