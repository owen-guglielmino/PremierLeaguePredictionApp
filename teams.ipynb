```python
from lxml import html
import requests
import pandas as pd
```

```python

url = 'https://fbref.com/en/comps/9/Premier-League-Stats'

stuff = pd.read_html(url, header=0)
df = stuff[0]
raw = df.reset_index(drop=True)
teams = raw
```

```python
teams = teams[['Rk', 'Squad', 'xGD/90']]
teams
```

```python
teams['xG'] = [2.12, 2.19, 1.91, 1.65, 1.81, 1.49, 1.14, 1.51, 1.72, 1.19, 1.51, 1.38, 1.42, 1.12, 1.39, 1.48, .95, 1.13, .94, .79]
teams
```

```python
teams = teams.to_csv('teams.csv', index=False)
```

```python
home = 'Burnley'
away = 'Luton Town'

home_feats = teams[teams['Squad'] == home].reset_index()
away_feats = teams[teams['Squad'] == away].reset_index()

home_feats = home_feats.drop(['index', 'Squad'], axis=1)
away_feats = away_feats.drop(['index', 'Squad'], axis=1)
home_feats
```

```python
home_feats = home_feats.loc[0, :].values.flatten().tolist()
away_feats = away_feats.loc[0, :].values.flatten().tolist()
```

```python
for i in range(len(home_feats)):
    home_feats.append(away_feats[i])
```

```python
order = [2, 5, 1, 4, 0, 3]
final_features = [home_feats[i] for i in order]
final_features
```
